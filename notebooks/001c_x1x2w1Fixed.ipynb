```python
# Definción de funciones
def data(x,w,b):
    return np.dot(x,w)+b
    
def sigmoide(z):
    return (1)/(1+np.exp(-z))

def sigmoide_2d(x1,x2,w1,w2,b):
    return (1)/(1+np.exp( -(x1*w1 + x2*w2 + b) ))

def cost(h,Y):
    return (h-Y)**2

def gradient(y,Y):
    return 2*(y-Y)*y*(1-y)
```

```python
x1 = 1
x2 = 2
w1 = 0.5
b  = 0

for i in np.arange(0,20,0.000001):
    w2     = i
    result = sigmoide_2d(x1,x2,w1,w2,b)
    if (abs(result - 0.71))<0.00001:
        print(w2,result)
        break
```
