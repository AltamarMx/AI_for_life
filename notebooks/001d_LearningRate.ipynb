```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
%matplotlib notebook
```

```python
# Definción de funciones
def data(x,w,b):
    space = 0
    for i in np.arange(len(w)):
        a     = x[i]*w[i]
        space = space + a
    return space
    
def sigmoide(z):
    return (1)/(1+np.exp(-z))

def cost(h,Y):
    return (h-Y)**2

def gradient(y,Y):
    return 2*(y-Y)*y*(1-y)
```

```python
epoch = 500
LR    = 0.01
x     = [3,2]
w     = [-1,-0.1]
b     = 0
C     = []
wg    = []
# making the lr to obtain the correct answer Y
for i in range(epoch):
    H = sigmoide(data(x,w,b))
    c = cost(H,0.71)
    w = w - LR*gradient(H,0.71)
    C.append(c)
    wg.append(w)
# making a list to iterate and graph later n points 
it = []
for i in np.linspace(0,len(C)-1,7):
    it.append(round(i))
```

```python
for i in it:
    print(wg[i][0], wg[i][1], C[i])
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
W1      = np.linspace(-5,5,100)
W2      = np.linspace(-5,5,100)
W11,W22 = np.meshgrid(W1,W2)
W       = [W11,W22]
X       = [1,5]
S = cost(sigmoide(data(X,W,b=0)),0.71)

# Plot the surface.
surf = ax.plot_surface(W11, W22, S, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False,alpha=0.2)
#ax.scatter(w1,w2,s)
for i in it:
    ax.scatter(wg[i][0], wg[i][1], C[i])

ax.view_init(30, 10)
plt.show()
```
